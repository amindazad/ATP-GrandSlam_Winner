```
# import data sets 
import pandas as pd 
import numpy as np

match_scores_91_16 = pd.read_csv('https://raw.githubusercontent.com/amindazad/ATP-GrandSlam_Winner/master/match_scores_1991-2016_unindexed.csv')
match_scores_17 = pd.read_csv('https://raw.githubusercontent.com/amindazad/ATP-GrandSlam_Winner/master/match_scores_2017_unindexed.csv')
match_stats_91_16 = pd.read_csv('https://raw.githubusercontent.com/amindazad/ATP-GrandSlam_Winner/master/match_stats_1991-2016_unindexed.csv')
match_stats_17 = pd.read_csv('https://raw.githubusercontent.com/amindazad/ATP-GrandSlam_Winner/master/match_stats_2017_unindexed.csv')
tournaments = pd.read_csv('https://raw.githubusercontent.com/amindazad/ATP-GrandSlam_Winner/master/tournaments_1877-2017_unindexed.csv')

match_scores_91_16.shape, match_scores_17.shape , match_stats_91_16.shape, match_stats_17.shape, tournaments.shape
```

```
# Concatenate and Merge into one dataset

#Contcatenate match_scores_91_16 and match_scores_17 by rows
match_scores = pd.concat([match_scores_91_16, match_scores_17])
#Contcatenate match_stats_91_16 and match_stats_17 by rows
match_stats = pd.concat([match_stats_91_16, match_stats_17])
#Merge using match_id
match_data = pd.merge(match_scores, match_stats, on='match_id', how='inner')
match_data.shape
```

```
# Drop the columns we dont need 
match_data = match_data.drop(columns=['tourney_url_suffix', 'match_stats_url_suffix_x', 'tourney_order_y', 'match_stats_url_suffix_y'])
# Create a new column indicating the year tornoment happened and convert it to datetime
match_data['tournament_year'] = match_data['tourney_year_id'].str[:4]
match_data['tournament_year'] = pd.to_datetime(match_data['tournament_year'], infer_datetime_format=True).dt.year
```

```
# Moving on to the tournaments dataset
# Convert tourney_dates to datetime format
tournaments['tourney_dates'] = pd.to_datetime(tournaments['tourney_dates'], infer_datetime_format=True)
# Filter out 1991 to 2017
tournaments = tournaments[tournaments['tourney_dates']>'1991-01-01']
tournaments
```

```
tournaments['tourney_year_id'] = tournaments[['tourney_year', 'tourney_id']].agg('-'.join, axis=1)
```

```
tourney_dates = tournaments['tourney_dates']
tourney_name = tournaments['tourney_name']
fin_commit = tournaments['tourney_fin_commit']
winner = tournaments['singles_winner_name']
slug = tournaments["tourney_slug"]
```

```
new_df = pd.DataFrame({
    "tourney_dates": tourney_dates,
    "tourney_name" : tourney_name,
    "fin_commit" : fin_commit,
    "winner" : winner,
    "tourney_slug" : slug
})
```
